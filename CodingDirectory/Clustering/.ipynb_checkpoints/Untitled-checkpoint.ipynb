```python
def topics_to_service(df_res,df_clus,col_name="ssdsLemma"):
    df_res["Topics"]=df_res.apply(lambda row: df_clus.iloc[df_res[f"{col_name}_cluster"].iloc[row.name],2], axis=1)
    if 'categorized' in df_clus:
        df_res["categorized"]=df_res.apply(lambda row: df_clus.iloc[df_res[f"{col_name}_cluster"].iloc[row.name],3], axis=1)
    return df_res
```

```python
def group_by_cluster(df,col_name="ssdsLemma"):
    df_work = df[[f"{col_name}_processed",f"{col_name}_cluster"]]
    df_work=df_work.groupby([f"{col_name}_cluster"])[f"{col_name}_processed"].apply(lambda x: ' '.join(x)).reset_index()
    return df_work
```

```python
def return_topics(df,n=1,col_name="ssdsLemma"):
    vectorizer=TfidfVectorizer()
    rows = vectorizer.fit_transform(df[f"{col_name}_processed"].tolist()).toarray()
    features=vectorizer.get_feature_names()
    i=0
    topics=[]
    while i<len(rows):
        top_feats=[]
        topn_ids = np.argsort(rows[i])[::-1][:n]
        top_feats = [(features[j]) for j in topn_ids]
        topics.append(top_feats)
        i+=1
    df['Topics']=topics
    return df
```

```python
def categorize_text(df,key,endpoint="https://berlinbobbi.cognitiveservices.azure.com/",col_name="ssdsLemma"):
    client=authenticate_azure_client(key,endpoint)
    topic=[None]*len(df)
    i=0
    while i<len(df):            
        documents=[df[f"{col_name}_processed"].iloc[i][0:5119]]    
        result =client.recognize_entities(documents)[0]
        summary=[]       
        for entity in result.entities:
            categorized=[entity.text,entity.category,round(entity.confidence_score,2),entity.offset]
            summary.append(categorized)
        topic[i]=(sorted(summary,key=itemgetter(3),reverse=False))
        i+=1
    df['categorized']=topic
    return df
```

```python
def authenticate_azure_client(key,endpoint="https://berlinbobbi.cognitiveservices.azure.com/"):
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
            endpoint=endpoint, 
            credential=ta_credential)
    return text_analytics_client
```
