```python
def aggregate_questions(df):
        #df = df.groupby(["dialogId", "question",'delta_rank','delta_nResult','nResult_deduction'])["t"].agg('max').to_frame('turns').reset_index()
        df = df.groupby([ "question"]).agg(Dialog_Count=('dialogId','count'),Services_Count=('name','nunique'),Min_Reduction=('nResult_reduction', 'min'),Mean_Reduction=('nResult_reduction', 'mean'), Max_Reduction=('nResult_reduction', 'max'),Mean_dRank=('delta_Rank','mean'))
        return df
```
