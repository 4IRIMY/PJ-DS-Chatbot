```python
def findCorrectAnswer(chatbot,targetService):
    
    # Find Target Cluster
    clusteredColumn = "cluster_label"
    service = chatbot.df_current.loc[chatbot.df_current["id"]==str(targetService)]
    if 'cluster_label' not in service.columns:
        return None
    
    if len(service[clusteredColumn].values) == 0:
        return None

    targetCluster = service[clusteredColumn].values[0]
    
    if targetCluster == chatbot.max_cluster_label_current:
        return True

    else:
        return False

def getRank(chatbot, targetService):
    temp = chatbot.df_current.reset_index()
    try:
        rank = temp.index[temp["id"]==str(targetService)].values[0]
    except IndexError as e:
        return float("NaN")
    return rank + 1
```

Logging

```python
class Logger:
    
    def __init__(self, eval_df):
        self.liste = []
        self.eval_df = eval_df
        self.chatbot = None
    
    def getDataFrame(self): # TODO
        pass
    
    def setChatbot(self, chatbot):
        self.chatbot = chatbot
    
    def log(self, ID, t, row, answer = None, initialQuery = False): 
        target_service = row["documentId"]
        #print(ID)
        #print(t)
        #print(answer)
        
        dialogId = row["dialogId"]
        file = row["file"]
        
        service_name = row["name"]
        query = row["initialQuestion"]
        if initialQuery==True:
            nResults = len(self.chatbot.df_current.index)
            rank = getRank(self.chatbot,target_service)# getRank(target_service)
            
            row = (file, dialogId, ID, t, service_name, query, "initialQuery", None, rank, nResults)
        else:
            if answer is not None:

                
                nResults = len(self.chatbot.df_current.index)
                question = self.chatbot.generateQuestion()
                answer = answer

                rank = getRank(self.chatbot,target_service)# getRank(target_service)

                row = (file, dialogId, ID, t,  service_name, query, question, answer, rank, nResults)
            else:
                row = (file, dialogId, ID, t, service_name, query, None, None, None, None)
        self.liste.append(row)
        #print(self.liste)
```

```python
lg = Logger(eval_df = eval)  

import os
import time

iterator = eval.iterrows()

if os.path.exists('./service_clus_dbscn_n1 -Silvio.csv'):
    seen = pd.read_csv('service_clus_dbscn_n1 -Silvio.csv', sep="~")
    seen = max(seen["ID"], key = lambda k: int(k) if k else 0)
    seen += 1
    for i in range(seen):
        next(iterator)
    
    print(f"Continuing at ID: {seen}")
else:
    with open('./service_clus_dbscn_n1 -Silvio.csv', "a+", encoding='utf-8') as tests:
        tests.write("file~dialogId~ID~t~name~query~question~answer~rank~nResult\n")
        tests.close()
    seen = 0
    
with open('./service_clus_dbscn_n1 -Silvio.csv', "a+", encoding='utf-8') as tests:
    for (ID, row) in iterator:
        
        #print progress
        print(ID)
        trace = ""
        #initialise arguments
        query = row["initialQuestion"]
        target_service = row["documentId"]
        t = 0
        
        # Chatbot initialisieren
        try: # 1.3
            chatbot =  ServiceClustering(query, dbscan, max_results=1)
            
        # Fehler überspringen: eg. kein solr output
        except Exception as e: 
            if(str(e) == 'response'): # TODO ??
                print(e)
                continue
            elif(str(e) == 'no solr output'):
                print(e)
                continue
            elif(str(e) == 'cluster_label'):
                print(e)
                continue
            else:
                print(e)
                continue
                
        lg.setChatbot(chatbot)
    
        answer = findCorrectAnswer(chatbot, target_service)
    
        # service nicht in resultset, überspringen
        if answer is None:
            continue
    
        lg.log(ID,t, row, answer, initialQuery = True)
        #file,dialogId,ID,t,name,query,question,answer,rank,nResult
        #tests.write(f"{row['file']},{row['dialogId']},{ID},{t},{row['name']},{query},{'initialQuery'},{''},{getRank(chatbot,target_service)},{len(chatbot.df_current.index)}\n")
        trace += f"{row['file']}~{row['dialogId']}~{ID}~{t}~{row['name']}~{query}~{'initialQuery'}~{''}~{getRank(chatbot,target_service)}~{len(chatbot.df_current.index)}\n"
        start_time = time.time()
        while not chatbot.isFinished():
    
            #log
            t += 1
    
            chatbot.refineResultset(answer)
            answer = findCorrectAnswer(chatbot, target_service)
    
            if (answer is None) or (time.time() - start_time > 300):
                print(f"Abbruch mit der Antwort {answer} nach {time.time() - start_time} Sekunden.")
                break
                
            lg.log(ID, t, row,answer)
            #file,dialogId,ID,t,name,query,question,answer,rank,nResult
            #tests.write(f"{row['file']},{row['dialogId']},{ID},{t},{row['name']},{query},{chatbot.generateQuestion()},{answer},{getRank(chatbot,target_service)},{len(chatbot.df_current.index)}\n")
            trace += f"{row['file']}~{row['dialogId']}~{ID}~{t}~{row['name']}~{query}~{chatbot.generateQuestion()}~{answer}~{getRank(chatbot,target_service)}~{len(chatbot.df_current.index)}\n"
        tests.write(trace)
    tests.close()
```
